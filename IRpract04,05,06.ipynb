```python
docs = {
    1: "kawasaki bmw toyota",
    2: "kawasaki bmw",
    3: "toyota bmw",
    4: "mitshubishi kawasaki"
}

query = input("Enter query: ").lower()

print("Matching Documents:")

for i in docs:
    text = docs[i]

    if "and" in query:
        w1 = query.split(" and ")[0]
        w2 = query.split(" and ")[1]
        if w1 in text and w2 in text:
            print(i)

    elif "or" in query:
        w1 = query.split(" or ")[0]
        w2 = query.split(" or ")[1]
        if w1 in text or w2 in text:
            print(i)

    elif "not" in query:
        w = query.split("not ")[1]
        if w not in text:
            print(i)

    else:
        if query in text:
            print(i)
```

```python
import numpy as np
import math
doc1 = "artificial intelligence is the future"
doc2 = "future technology uses artificial intelligence"

words1 = doc1.lower().split()
words2 = doc2.lower().split()

vocab = list(set(words1 + words2))

freq1 = dict.fromkeys(vocab, 0)
freq2 = dict.fromkeys(vocab, 0)

for word in words1:
    freq1[word] += 1

for word in words2:
    freq2[word] += 1

# Convert to vectors
vec1 = list(freq1.values())
vec2 = list(freq2.values())

# Cosine similarity
def cosine_similarity(v1, v2):
    dot = np.dot(v1, v2)
    mag1 = math.sqrt(np.dot(v1, v1))
    mag2 = math.sqrt(np.dot(v2, v2))
    return dot / (mag1 * mag2)

print("Vocabulary:", vocab)
print("Vector 1:", vec1)
print("Vector 2:", vec2)
print("Cosine Similarity:", cosine_similarity(vec1, vec2))
```

```python
import math

# Documents and Query
D1 = "information retrieval system uses cosine similarity"
D2 = "retrieval system uses jaccard similarity measure"
D3 = "information system ranking using vector space model"
Q  = "information retrieval similarity"

docs = {"D1": D1, "D2": D2, "D3": D3, "Q": Q}

# Create vocabulary
words = sorted(set(" ".join(docs.values()).split()))

# ---------------------------
# 1. Binary Term Incidence Matrix
# ---------------------------
print("\nBINARY TERM INCIDENCE MATRIX\n")

print("{:<5}".format("Doc"), end="")
for w in words:
    print("{:<12}".format(w), end="")
print()

for name, text in docs.items():
    print("{:<5}".format(name), end="")
    for w in words:
        value = 1 if w in text.split() else 0
        print("{:<12}".format(value), end="")
    print()

# ---------------------------
# 2. Cosine Similarity
# ---------------------------
def cosine(d, q):
    v1 = [d.split().count(w) for w in words]
    v2 = [q.split().count(w) for w in words]

    dot = sum(a*b for a,b in zip(v1,v2))
    norm1 = math.sqrt(sum(a*a for a in v1))
    norm2 = math.sqrt(sum(b*b for b in v2))

    if norm1 == 0 or norm2 == 0:
        return 0
    return dot/(norm1*norm2)

# ---------------------------
# 3. Jaccard Similarity
# ---------------------------
def jaccard(d, q):
    s1 = set(d.split())
    s2 = set(q.split())
    return len(s1 & s2) / len(s1 | s2)

# ---------------------------
# 4. Final Combined Table
# ---------------------------
print("\nFINAL SIMILARITY TABLE\n")
print("{:<10}{:<15}{:<15}".format("Document","Cosine","Jaccard"))

for name in ["D1","D2","D3"]:
    cos = round(cosine(docs[name], Q),4)
    jac = round(jaccard(docs[name], Q),4)
    print("{:<10}{:<15}{:<15}".format(name, cos, jac))
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
import pandas as pd
import numpy as np
documents = [
    "machine learning is very useful",
    "python is used for machine learning",
    "data science and machine learning",
    "python is popular in data science"
]

count_vectorizer = CountVectorizer()
count_matrix = count_vectorizer.fit_transform(documents)

tf_df = pd.DataFrame(count_matrix.toarray(), columns=count_vectorizer.get_feature_names_out())

print("Term Frequency (TF) Matrix (Raw Counts):\n")
print(tf_df)
print("\n" + "-"*50 + "\n")

tfidf_vectorizer_for_idf = TfidfVectorizer(use_idf=True, smooth_idf=True)
tfidf_vectorizer_for_idf.fit(documents)
feature_names = tfidf_vectorizer_for_idf.get_feature_names_out()
idf_values = tfidf_vectorizer_for_idf.idf_
idf_df = pd.DataFrame([idf_values], columns=feature_names, index=['IDF'])

print("Inverse Document Frequency (IDF) Values:\n")
print(idf_df)
print("\n" + "-"*50 + "\n")
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vectorizer.get_feature_names_out())
print("TF-IDF Matrix:\n")
print(tfidf_df)
```
