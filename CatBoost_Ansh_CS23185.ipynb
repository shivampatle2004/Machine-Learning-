```python
import pandas as pd
```

```python
path="/content/drive/MyDrive/ML/BreastCancer.csv"
data=pd.read_csv(path)
```

```python
print(data.head())
print(data.shape)
```

```python
print(data.info())
print(data.isnull().sum())
```

```python
data = data.drop(['id', 'Unnamed: 32'], axis=1)
```

```python
data['diagnosis'] = data['diagnosis'].map({'M':1, 'B':0})
```

```python
X = data.drop('diagnosis', axis=1)
y = data['diagnosis']
```

```python
!pip install catboost
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix
from catboost import CatBoostClassifier
import seaborn as sns
import matplotlib.pyplot as plt

for split in [0.2, 0.3]:

    print("\n==============================")
    print(f"CatBoost Split: {int((1-split)*100)}-{int(split*100)}")
    print("==============================")

    # Split
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=split, random_state=42
    )

    # Model
    cb_model = CatBoostClassifier(verbose=0)
    cb_model.fit(X_train, y_train)

    # Prediction
    y_pred = cb_model.predict(X_test)

    # Accuracy
    acc = accuracy_score(y_test, y_pred)
    print("Accuracy:", acc)

    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)

    plt.figure()
    sns.heatmap(cm, annot=True, fmt='d')
    plt.title(f"CatBoost Confusion Matrix ({int((1-split)*100)}-{int(split*100)})")
    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()
```

```python
importance = cb_model.get_feature_importance()

plt.figure()
plt.barh(X.columns, importance)
plt.title("CatBoost Feature Importance")
plt.show()
```
